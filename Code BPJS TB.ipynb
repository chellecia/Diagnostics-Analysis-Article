#### Visualisasi

```python
# Menghitung frekuensi dari setiap kode dalam kolom 'Kode Diagnosis ICD-10 (3 Digit)'
top_codes = filtered_data_TB['Kode Diagnosis ICD-10 (3 Digit)'].value_counts().head(5)

# Mengganti label kode diagnosis dengan deskripsi lengkap
labels = [
    'E11: Diabetes Mellitus Tipe 2',
    'I10: Hipertensi Esensial (primer)',
    'E14: Diabetes Mellitus Tipe 2 (tanpa komplikasi)',
    'Z30: Kontrasepsi dan pengaturan keluarga',
    'O80: Persalinan tunggal, normal'
]

# Warna khusus untuk setiap sektor
colors = ['#D56C42', '#FF7F50', '#F4A300', '#E64A19', '#D32F2F']

# Menambahkan efek explode untuk bagian dengan frekuensi tertinggi
explode = [0.1 if i == 0 else 0 for i in range(len(top_codes))]

# Membuat pie chart
plt.figure(figsize=(10, 8))
plt.pie(
    top_codes, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    explode=explode,
    shadow=True,   # Menambahkan bayangan
    wedgeprops={'edgecolor': 'black'}  # Membuat garis pembatas antar sektor
)

# Menambahkan judul
plt.title('Distribusi 5 Kode Diagnosis ICD-10 Teratas', fontsize=16, fontweight='bold')

# Menampilkan plot
plt.show()
```

```python

# Mengonversi "Tanggal Lahir Peserta" menjadi usia
filtered_data_TB['Tanggal Lahir Peserta'] = pd.to_datetime(filtered_data_TB['Tanggal Lahir Peserta'])
current_year = pd.to_datetime('today').year
filtered_data_TB['Usia'] = current_year - filtered_data_TB['Tanggal Lahir Peserta'].dt.year

# Mengelompokkan usia ke dalam kategori
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '60+']
filtered_data_TB['Kelompok Usia'] = pd.cut(filtered_data_TB['Usia'], bins=bins, labels=labels, right=False)

# Menghitung distribusi kode diagnosis ICD-10 berdasarkan kelompok usia
age_group_distribution = filtered_data_TB.groupby(['Kelompok Usia', 'Kode Diagnosis ICD-10 (3 Digit)']).size().unstack()

# Memilih 5 kode diagnosis teratas berdasarkan total jumlah kasus
top_codes = age_group_distribution.sum().nlargest(5)
age_group_distribution_top = age_group_distribution[top_codes.index]

# Menggunakan colormap 'Oranges' dengan nilai tinggi artinya warna lebih gelap
cmap = plt.cm.Oranges_r  # .r membalikkan colormap

# Menampilkan stacked bar plot
plt.figure(figsize=(10, 8))
age_group_distribution_top.plot(kind='bar', stacked=True, figsize=(10, 8), colormap=cmap, edgecolor='black')

# Menambahkan grid, judul, dan label
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid garis putus-putus
plt.title('Distribusi Kode Diagnosis ICD-10 Teratas Berdasarkan Kelompok Usia', fontsize=16, fontweight='bold')
plt.xlabel('Kelompok Usia', fontsize=14)
plt.ylabel('Jumlah Kasus', fontsize=14)
plt.xticks(rotation=45)

# Menambahkan annotasi pada setiap bar untuk menunjukkan jumlah kasus
for i, bar_container in enumerate(plt.gca().containers):
    for bar in bar_container:
        height = bar.get_height()
        if height > 0:
            plt.gca().text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{int(height)}',
                ha='center',
                va='center',
                fontsize=10,
                color='white'
            )

# Menampilkan plot dengan layout rapi
plt.tight_layout()
plt.show()
```
